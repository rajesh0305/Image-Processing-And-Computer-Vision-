```python
from skimage import color, data

#Load grayscale image
image = color.rgb2gray(data.coffee())
```

```python
import numpy as np
#Define 2x2 max pooling function
def max_pooling(feature_map, size = 2, stride=2):
    output_shape = (
        (feature_map.shape[0]-size) // stride + 1,
        (feature_map.shape[1]-size) // stride + 1
    )
    pooled = np.zeros(output_shape)
    for i in range(0, output_shape[0]):
        for j in range(0, output_shape[1]):
            region = feature_map[i*stride:i*stride+size, j*stride:j*stride+size]
            pooled[i, j] = np.max(region) # Corrected line: changed * to =

    return pooled
```

```python
from scipy.signal import convolve2d

# Define a simple kernel for convolution (e.g., edge detection)
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Perform convolution. The 'image' variable is already loaded.
# 'mode="valid"' ensures the output is only where the kernel fully overlaps the input.
convolved = convolve2d(image, kernel, mode='valid')
```

```python
#Apply MaxPooling
pooled = max_pooling(feature_map=convolved, size=2, stride=2)
```

```python
# Showing The Result
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,6)) # Adjust figure size for 3 plots
plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Convolved Image')
plt.imshow(convolved, cmap='gray')
plt.axis('off') # Also turn off axis for convolved for consistency

plt.subplot(1,3,3)
plt.title('Pooled Feature Map')
plt.imshow(pooled, cmap='gray')
plt.axis('off') # Turn off axis for pooled

plt.tight_layout()
plt.show()
```

### 1. Load Image
First, we'll load a sample image and convert it to grayscale using `skimage`.

```python
from skimage import color, data

# Load grayscale image
image = color.rgb2gray(data.coffee())
```

### 2. Define Max Pooling Function
Next, we define the `max_pooling` function. I've corrected the assignment operation here to ensure accurate results.

```python
import numpy as np

#Define 2x2 max pooling function
def max_pooling(feature_map, size = 2, stride=2):
    output_shape = (
        (feature_map.shape[0]-size) // stride + 1,
        (feature_map.shape[1]-size) // stride + 1
    )
    pooled = np.zeros(output_shape)
    for i in range(0, output_shape[0]):
        for j in range(0, output_shape[1]):
            region = feature_map[i*stride:i*stride+size, j*stride:j*stride+size]
            pooled[i, j] = np.max(region) # Corrected line for accurate assignment

    return pooled
```

### 3. Perform Convolution
Now, we'll apply a simple convolution kernel to the grayscale image using `scipy.signal.convolve2d`.

```python
from scipy.signal import convolve2d

# Define a simple kernel for convolution (e.g., edge detection)
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Perform convolution. The 'image' variable is already loaded.
# 'mode="valid"' ensures the output is only where the kernel fully overlaps the input.
convolved = convolve2d(image, kernel, mode='valid')
```

### 4. Apply Max Pooling
After convolution, we'll apply our `max_pooling` function to the convolved image.

```python
# Apply MaxPooling
pooled = max_pooling(feature_map=convolved, size=2, stride=2)
```

### 5. Visualize Results
Finally, let's visualize the original image, the convolved image, and the pooled feature map to see the effects of each step.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,6)) # Adjust figure size for 3 plots
plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Convolved Image')
plt.imshow(convolved, cmap='gray')
plt.axis('off') # Also turn off axis for convolved for consistency

plt.subplot(1,3,3)
plt.title('Pooled Feature Map')
plt.imshow(pooled, cmap='gray')
plt.axis('off') # Turn off axis for pooled

plt.tight_layout()
plt.show()
```
